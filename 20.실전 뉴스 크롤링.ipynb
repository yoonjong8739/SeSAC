실전 뉴스 크롤링
- 아래 URL의 매체의 어제 일자 모든 기사를 제목과 분문을 분류하여 크롤링하시오.
- https://news.naver.com/main/list.naver?mode=LSD&mpde=LPOD&mid=sec&oid=003

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
for i in range(54, 0, -1):  # 오름차순 정렬
    URL = f"https://news.naver.com/main/list.naver?mode=LSD&mpde=LPOD&mid=sec&oid=003&date=20251102&page={i}"

    response = requests.get(URL)
    news_soup = BeautifulSoup(response.text, "html.parser")
    second_news_list = news_soup.select("#main_content ul dt.photo > a")

    #print(second_news_list)
    #print(len(second_news_list)) # 12

    for idx in range(len(second_news_list)):
        #second_news_title = second_news_list[idx].select_one("img")["alt"]
        response2 = requests.get(second_news_list[idx]["href"])    
        detail_news_soup = BeautifulSoup(response2.text, "html.parser")
        
        title = None
        for sel in ["#title_area", ".ArticleHead_article_title__qh8GV"]:
            tag = detail_news_soup.select_one(sel)
            if tag:
                title = tag.get_text(strip=True)
                break
            
        article = None
        for sel in ["#newsct_article > #dic_area", "div._article_content"]:
            tag = detail_news_soup.select_one(sel)
            if tag:
                article = tag.get_text(strip=True)
                break
        print(title)
        print(article)
        print()
```
