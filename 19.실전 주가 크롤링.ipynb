```python
# 종목명과 현재가 크롤링
import requests
from bs4 import BeautifulSoup

URL = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(URL)

finance_soup = BeautifulSoup(response.text, "html.parser")
total_data = finance_soup.select(".type_2 tr")

for item in total_data:
    if len(item.select("td")) != 12:
        continue
    name = item.select("td")[1].text
    current_price = item.select("td")[2].text
    print(name, current_price)
```

```python
# 전일대비 상승한 항목만 종목명, 현재가, 전일비를 크롤링
import requests
from bs4 import BeautifulSoup

URL = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(URL)

finance_soup = BeautifulSoup(response.text, "html.parser")
total_data = finance_soup.select(".type_2 tr")

for item in total_data:
    if len(item.select("td")) != 12:
        continue
    status = [item.select("td")[3].text]
    if status[0].replace("\n","").replace("\t","")[:2] == "상승" :
        name = item.select("td")[1].text
        current_price = item.select("td")[2].text
        diff_rate = item.select("td")[3].text.replace("\n","").replace("\t","")[2:]
        print(name, current_price, diff_rate)
```
