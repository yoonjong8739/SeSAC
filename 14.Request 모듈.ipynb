Requests 모듈

```python
import requests

URL = "http://httpbin.org/get"

response = requests.get(URL)  # 서버에 GET 요청

print(response.status_code)
print(response.text)
```

```python
URL2 = "http://naver.com"

response2 = requests.get(URL2)

print(response2.status_code)
print(response2.text)
```

```python
URL = "http://httpbin.org/post"
response3 = requests.post(URL)  # 신규 데이터를 저장할 때 POST 요청

print(response3.status_code)
print(response3.text)
```

```python
URL = "http://httpbin.org/put"  # 서버에 저장된 데이터의 속성들을 모두 변경할 때 PUT 요청

response4 = requests.put(URL)
print(response4.status_code)
print(response4.text)
```

```python
URL = "http://httpbin.org/delete"  # 데이터를 삭제할 때 DELETE 요청

response5 = requests.delete(URL)
print(response5.status_code)
print(response5.text)
```

params

```python
import requests

URL = "http://httpbin.org/get"

params = {"data1": "value1", "data2": "value2"}
response = requests.get(URL, params=params)

print(response.status_code)
print(response.text)
```

header

```python
import requests

URL = "http://httpbin.org/get"

params = {"data1": "value1", "data2": "value2"}
headers = {"Content-Type": "application/json; charset=utf-8"}
response = requests.get(URL, params=params, headers=headers)

print(response.status_code)
print(response.text)
```

Body

```python
import requests

URL = "http://httpbin.org/post"

params = {"data1": "value1", "data2": "value2"}
headers = {"Content-Type": "application/json; charset=utf-8"}

data = {"key1": "value1", "key2": "value2"}
response = requests.get(URL, params=params, headers=headers, data=data)

print(response.status_code)
print(response.text)
```
